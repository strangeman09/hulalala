```python
import random 
random.seed(42)
def make_arr(n):
    weight=[]
    
    
    i=1
    len=n
    while n>0:
        rand_int=random.randint(0,n)
        weight.append(rand_int)
        
        
        n=n-1
        i=i+1
 
    return weight
```

```python
def find_min_max(a_list):

    if a_list:
        head, *tail = a_list

        if tail:
            minimum, maximum = find_min_max(tail)

            return [head, minimum][minimum < head], [head, maximum][maximum > head]

        return head, head

    return a_list
```

```python
from datetime import datetime
m=int(input('no.of arrays'))
length1=m
time_taken=[]
lengths=[]
while m>0:
    size=random.randint(0,length1)
    lengths.append(size)
    weight=make_arr(size)
    start=datetime.now()
    
    find_min_max(weight)
    end=datetime.now()
    time_taken.append((end-start).total_seconds())
    m=m-1
```

```python
import matplotlib.pyplot as plt
plt.plot(lengths,time_taken)
plt.ylabel('time taken')
plt.xlabel('size of the array')
```
