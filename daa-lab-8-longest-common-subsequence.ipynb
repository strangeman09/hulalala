```python
import random 
random.seed(42)
def make_arr(n):
    weight=[]
    
    
    i=1
    len=n
    while n>0:
        rand_int=random.randint(0,n)
        weight.append(str(rand_int))
        
        
        n=n-1
        i=i+1
 
    return weight
```

```python
def lcs(s1 , s2):
    m, n = len(s1), len(s2)
    prev, cur = [0]*(n+1), [0]*(n+1)
    for i in range(1, m+1):
        for j in range(1, n+1):
            if s1[i-1] == s2[j-1]:
                cur[j] = 1 + prev[j-1]
            else:
                if cur[j-1] > prev[j]:
                    cur[j] = cur[j-1]
                else:
                    cur[j] = prev[j]
        cur, prev = prev, cur
    return prev[n]


```

```python
from datetime import datetime
m=int(input('no.of arrays'))
length1=m
time_taken=[]
lengths=[]
while m>0:
    size=random.randint(0,length1)
    lengths.append(size)
    weight1=make_arr(size)
    weight2=make_arr(size)
    start=datetime.now()
    
    lcs(weight1,weight2)
    end=datetime.now()
    time_taken.append((end-start).total_seconds())
    m=m-1
```

```python
import matplotlib.pyplot as plt
plt.plot(lengths,time_taken)
plt.ylabel('time taken')
plt.xlabel('size of the array')
```
