```python
def knapsack(W,arr):
    profit=0
    arr.sort(key=lambda x: (x.value/x.weight), reverse=True)
    for Item in arr:
        if Item.weight <= W:
            W-=Item.weight
            profit+=Item.value
        else:
            profit+=Item.value*W/Item.weight
            break
    return profit
```

```python
class Item:
    def __init__(self,value,weight):
        self.value = value
        self.weight = weight
```

```python
import random
from datetime import datetime
def samples(n):
    result=[]
    sizes=[]
    for i in range(0,n):
        weight=100*n*n
        res=[]
        p=[]
        w=[]
        
        for j in range(0,n//2):
            p.append(random.randint(n,100*n))
            w.append(random.randint(n,10*n))
            res.append(Item(p[j],w[j]))
            start=datetime.now()
            knapsack(weight,res)
            end=datetime.now()
            sizes.append(2*j)
            result.append((end-start).total_seconds())
    return result,sizes
```

```python
import matplotlib.pyplot as plt
n=int(input())
ans,size=samples(n)
plt.plot(size,ans)
```
