```python
import heapq
def JobScheduling(arr,length):
    n = length
 
    # arr[i][0] = job_id, arr[i][1] = deadline, arr[i][2] = profit
 
    
    arr.sort(key=lambda x: x[1])
 
   
    result = []
    maxHeap = []
 
   
    for i in range(n - 1, -1, -1):
 
        # calculate slots between two deadlines
        if i == 0:
            slots_available = arr[i][1]
        else:
            slots_available = arr[i][1] - arr[i - 1][1]
 
        
        heapq.heappush(maxHeap, (-arr[i][2], arr[i][1], arr[i][0]))
 
        while slots_available and maxHeap:
 
            
            profit, deadline, job_id = heapq.heappop(maxHeap)
 
            
            slots_available -= 1
 
            
            result.append([job_id, deadline])
 
 
    result.sort(key=lambda x: x[1])
 
```

```python
import random 
random.seed(42)
def make_arr(n):
    arr=[]
    i=1
    len=n
    while n>0:
        rand_int=random.randint(0,n)
        arr.append([i,rand_int,10*rand_int])
        n=n-1
        i=i+1
    return arr
```

```python
from datetime import datetime
m=int(input('no.of arrays'))
len=m
time_taken=[]
lengths=[]
while m>0:
    size=random.randint(0,len)
    lengths.append(size)
    arr=make_arr(size)
    start=datetime.now()
    JobScheduling(arr,size)
    end=datetime.now()
    time_taken.append((end-start).total_seconds())
    m=m-1
```

```python
import matplotlib.pyplot as plt
plt.plot(lengths,time_taken)
plt.ylabel('time taken')
plt.xlabel('size of the array')
```
