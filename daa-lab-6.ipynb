```python
def knapSack(W, wt, val, n):
    K = [[0 for x in range(W + 1)] for x in range(n + 1)]
 
    
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                K[i][w] = 0
            elif wt[i-1] <= w:
                K[i][w] = max(val[i-1]
                          + K[i-1][w-wt[i-1]], 
                              K[i-1][w])
            else:
                K[i][w] = K[i-1][w]
 
    return K[n][W]
```

```python
import random 
random.seed(42)
def make_arr(n):
    weight=[]
    value=[]
    
    i=1
    len=n
    while n>0:
        rand_int=random.randint(0,n)
        weight.append(rand_int)
        value.append(10*rand_int)
        
        n=n-1
        i=i+1
    val=random.randint(0,n)
    return weight,value,val
```

```python
weight,value,val=make_arr(40)
```

```python
from datetime import datetime
m=int(input('no.of arrays'))
len=m
time_taken=[]
lengths=[]
while m>0:
    size=random.randint(0,len)
    lengths.append(size)
    weight,value,val=make_arr(size)
    start=datetime.now()
    knapSack(val, weight, value, m)
    end=datetime.now()
    time_taken.append((end-start).total_seconds())
    m=m-1
```

```python
import matplotlib.pyplot as plt
plt.plot(lengths,time_taken)
plt.ylabel('time taken')
plt.xlabel('size of the array')
```
