```python
import time
import matplotlib.pyplot as plt
import numpy as np

numbers=[]
for i in range(69,569):
    numbers.append(i)
list=[]

def fact(n):
    f=1
    if n==0:
        f=1
    elif n==1:
        f=1 
    else:
        for i in range(1,n + 1):
            f = f*i
    return f
    
for n in numbers:
    st = time.perf_counter_ns()
    fact(n)
    et = time.perf_counter_ns()
    elapsed_time = et - st
    list.append(elapsed_time) 
#print(list)

plt.plot(numbers,list)

plt.xlabel("number")
plt.ylabel("time taken")

```

```python
import time
import matplotlib.pyplot as plt
import numpy as np

numbers=[5,10,15,20,25]
mumbers=[25,30,35,40,45]
list=[]

def gcd(a, b):
   if a == b:
      return a
   elif a < b:
      return gcd(b, a)
   else:
      return gcd(b, a - b)


for n in numbers:
    st = time.time()
    for m in mumbers:
           gcd(n,m)
    et = time.time()
    elapsed_time = et - st
    list.append(elapsed_time)  
    
print(list)

plt.plot(numbers,list)

plt.xlabel("number")
plt.ylabel("time taken")





```
